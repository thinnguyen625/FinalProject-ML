### `import` mã nguồn

và khai báo một số biến cần dùng:

- `n_samples`: điều chỉnh số mẫu để chạy (phải nhỏ hơn kích thước của `val_Y` và `test_Y`)    

- `gammas`: nếu không muốn dùng các $\gamma$ (gamma) mặc định (`[0.0005, 0.005, 0.025, 0.05, 0.25, 0.5]`)

- `strategy`: nếu không muốn dùng chiến lược phân loại mặc định `ovo`

```python
import gzip
import numpy as np
from pickle import dump, load


def read_mnist(mnist_file):
    """
    Reads MNIST data.
    
    Parameters
    ----------
    mnist_file : string
        The name of the MNIST file (e.g., 'mnist.pkl.gz').
    
    Returns
    -------
    (train_X, train_Y, val_X, val_Y, test_X, test_Y) : tuple
        train_X : numpy array, shape (N=50000, d=784)
            Input vectors of the training set.
        train_Y: numpy array, shape (N=50000)
            Outputs of the training set.
        val_X : numpy array, shape (N=10000, d=784)
            Input vectors of the validation set.
        val_Y: numpy array, shape (N=10000)
            Outputs of the validation set.
        test_X : numpy array, shape (N=10000, d=784)
            Input vectors of the test set.
        test_Y: numpy array, shape (N=10000)
            Outputs of the test set.
    """
    f = gzip.open(mnist_file, 'rb')
    train_data, val_data, test_data = load(f, encoding='latin1')
    f.close()
    
    train_X, train_Y = train_data
    val_X, val_Y = val_data
    test_X, test_Y = test_data    
    
    return train_X, train_Y, val_X, val_Y, test_X, test_Y
```

### Lọc ra một DataFrame thể hiện thông số cho mỗi mô hình tốt nhất ứng với mỗi C

```python
import pandas as pd
from pickle import dump, load

def get_best_C(list_C):
    '''
    Get a DataFrame contains details of the best models for all the C
    Params:
    -----
        list_C: a list of the C params
    '''
    gamma, fit_time, tr_err, va_err, name = [], [], [], [], []
    
    for c in list_C:
        fname = 'rbf_' + str(c) + '_ovo'
        fpath = 'res/' + fname + '.csv'
        df = pd.read_csv(fpath)
        best_va_idx = df[df['best_va'] == True].index.values
        gamma.append(df.loc[best_va_idx, 'gamma'].values[0])
        fit_time.append(df.loc[best_va_idx, 'fit_time'].values[0])
        tr_err.append(df.loc[best_va_idx, 'tr_err'].values[0])
        va_err.append(df.loc[best_va_idx, 'va_err'].values[0])
        name.append(df.loc[best_va_idx, 'name'].values[0])
        
    best_C_data = {
        'C': list_C,
        'gamma': gamma,
        'fit_time': fit_time,
        'tr_err': tr_err,
        'va_err': va_err,
        'name': name
    }
    
    df_best_C = pd.DataFrame(data=best_C_data).set_index(['C'])
    return df_best_C

C = [0.001, 0.01, 0.1, 1.0, 5.0, 10.0, 50.0]
df_best_C = get_best_C(C)
df_best_C
```

### Mô hình tốt nhất là:

```python
minidx = df_best_C['va_err'].idxmin()
df_best_C[df_best_C.index == minidx]
```

### Giờ ta lấy mô hình tốt nhất vừa tìm được ở trên để dự đoán trên tập test

```python
best_name = df_best_C[df_best_C.index == minidx]['name'].values[0]
print('Mô hình tốt nhất là: %s' % best_name)
best_svc = load(open('res/' + best_name + '.pkl', 'rb'))
te_err = 1 - best_svc.score(test_X, test_Y)
print('Testing error: %f' % te_err)
```
